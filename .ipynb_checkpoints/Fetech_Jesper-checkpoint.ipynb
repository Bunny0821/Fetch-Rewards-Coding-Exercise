###  Second: Write queries that directly answer predetermined questions from a business stakeholder

```python
def load_json_lines(file_path):
    data = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            data.append(json.loads(line))
    return data

brands_path = 'brands.json'
receipts_path = 'receipts.json'
users_path = 'users.json'

brands_data = load_json_lines(brands_path)
receipts_data = load_json_lines(receipts_path)
users_data = load_json_lines(users_path)

brands_df = pd.DataFrame(brands_data)
receipts_df = pd.DataFrame(receipts_data)
users_df = pd.DataFrame(users_data)

receipts_list = []
items_list = []

for receipt in receipts_data:
    receipt_copy = receipt.copy()
    items = receipt_copy.pop('rewardsReceiptItemList', [])
    receipts_list.append(receipt_copy)
    
    for item in items:
        item['receipt_id'] = receipt['_id']['$oid']
        items_list.append(item)

receipts_df = pd.DataFrame(receipts_list)
items_df = pd.DataFrame(items_list)

if isinstance(users_df['_id'].iloc[0], dict) and '$oid' in users_df['_id'].iloc[0]:
    users_df['user_id'] = users_df['_id'].apply(lambda x: x['$oid'])
else:
    users_df['user_id'] = users_df['_id']
users_df = users_df.drop(columns=['_id'])
users_df['createdDate'] = users_df['createdDate'].apply(lambda x: pd.to_datetime(x['$date'], unit='ms') if isinstance(x, dict) and '$date' in x else x)
users_df['lastLogin'] = users_df['lastLogin'].apply(lambda x: pd.to_datetime(x['$date'], unit='ms') if isinstance(x, dict) and '$date' in x else x)

brands_df['brand_id'] = brands_df['_id'].apply(lambda x: x['$oid'] if isinstance(x, dict) and '$oid' in x else x)
brands_df = brands_df.drop(columns=['_id', 'cpg'])

receipts_df['receipt_id'] = receipts_df['_id'].apply(lambda x: x['$oid'] if isinstance(x, dict) and '$oid' in x else x)
receipts_df['user_id'] = receipts_df['userId']
receipts_df = receipts_df.drop(columns=['_id', 'userId'])
receipts_df['createDate'] = receipts_df['createDate'].apply(lambda x: pd.to_datetime(x['$date'], unit='ms') if isinstance(x, dict) and '$date' in x else x)
receipts_df['dateScanned'] = receipts_df['dateScanned'].apply(lambda x: pd.to_datetime(x['$date'], unit='ms') if isinstance(x, dict) and '$date' in x else x)
receipts_df['finishedDate'] = receipts_df['finishedDate'].apply(lambda x: pd.to_datetime(x['$date'], unit='ms') if isinstance(x, dict) and '$date' in x else x)
receipts_df['modifyDate'] = receipts_df['modifyDate'].apply(lambda x: pd.to_datetime(x['$date'], unit='ms') if isinstance(x, dict) and '$date' in x else x)
receipts_df['pointsAwardedDate'] = receipts_df['pointsAwardedDate'].apply(lambda x: pd.to_datetime(x['$date'], unit='ms') if isinstance(x, dict) and '$date' in x else x)
receipts_df['purchaseDate'] = receipts_df['purchaseDate'].apply(lambda x: pd.to_datetime(x['$date'], unit='ms') if isinstance(x, dict) and '$date' in x else x)

items_df['receipt_id'] = items_df['receipt_id'].apply(lambda x: x['$oid'] if isinstance(x, dict) and '$oid' in x else x)

print("Cleaned Users DataFrame:\n", users_df.head(5))
print("\nCleaned Brands DataFrame:\n", brands_df.head(5))
print("\nCleaned Receipts DataFrame:\n", receipts_df.head(5))
print("\nCleaned ReceiptItems DataFrame:\n", items_df.head(5))
```
